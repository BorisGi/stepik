Вася составил раскладку по продуктам на весь поход (она на листе "Раскладка") с указанием номера дня, названия продукта и его количества в граммах. Для каждого дня посчитайте 4 числа: суммарную калорийность и граммы белков, жиров и углеводов. Числа округлите до целых вниз и введите через пробел. Информация о каждом дне должна выводиться в отдельной строке.

```python
wb = xlrd.open_workbook('trekking3.xlsx') # читаю таблицу
sheet_1 = wb.sheet_by_index(0)
sheet_2 = wb.sheet_by_index(1)
```

```python
val_1 = [sheet_1.row_values(rownum) for rownum in range(sheet_1.nrows)]
val_2 = [sheet_2.row_values(rownum) for rownum in range(sheet_2.nrows)]
for i in val_1:
    if i[4] == '':
        i[4] = 0
```

```python
all_products = {} # создаю словарь с продуктами

for i in val_1:
    all_products[i[0]] = i[1:]
```

```python
calorific_value = {} # словарь для конечных расчетов

for i in val_2[1:]: # цикл по всем позициям
    
    temp = [] # временный список для расчетов

    if i[0] not in calorific_value: # проверяю есть ли такой день в словаре
        calorific_value[i[0]] = [0, 0, 0, 0] # создаю пустой список для дня
        
        for n in all_products[i[1]]:
            temp += [(i[2] * n) / 100]
         
    else:
        for n in all_products[i[1]]:
            temp += [(i[2] * n) / 100]
            
            
    for pl in enumerate(temp):
        calorific_value[i[0]][pl[0]] += pl[1]
```

```python
for k, v in calorific_value.items():
    for i in v:
        print(int(i), end=' ')
    print()
```
