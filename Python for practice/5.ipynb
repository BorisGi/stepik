```python
#!/usr/bin/env python3

print('<html>')
print('<body>')
print('<table>')
max_ = 10
for row in range(1, max_ + 1):
    print('<tr>')
    for column in range(1, max_ + 1):
        print('<td>', row * column, '</td>', sep='')
    print('</tr>')
print('</table>')
print('</body>')
print('</html>')
```

```python
#!/usr/bin/env python3

print('<html>')
print('<body>')
print('<table>')
n = 10
for row in range(1, n + 1):
    print('<tr>')
    for column in range(1, n + 1):
        print('<td>', '<a href="http://', row * column, '.ru">', row * column, '</a>', '</td>', sep='')
    print('</tr>')
print('</table>')
print('</body>')
print('</html>')
```
