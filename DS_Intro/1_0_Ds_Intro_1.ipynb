### 1.6 Группировка и агрегация ###

```python
(
    students_df.groupby('gender', as_index=False)
    .agg({'math_score': 'mean'})
    .rename(columns={'math_score': 'math_score_mean'})
)
```

```python
students_df.head(2)
```

```python
# мультииндексы

mean_scores = (
    students_df.groupby(['gender', 'race/ethnicity'])
    .agg({'math_score': 'mean', 'reading_score': 'mean', 'writing_score': 'mean'})
    .rename(columns={'math_score': 'math_score_mean',
                     'reading_score': 'reading_score_mean',
                     'writing_score': 'writing_score_mean'})
)
```

```python
# мультииндексы - составное обращение 

mean_scores.loc[('female', 'group A')]
```

```python
# мультииндексы создают сгруппированную серию, значения на пересечении

students_df.groupby(['gender', 'race/ethnicity']).math_score.nunique()
```

```python
(
    students_df.sort_values(['gender', 'math_score'], ascending=False)
    .groupby('gender')
    .head()
)
```

```python
# создание новых колонок

students_df['total_score'] = students_df.math_score + students_df.reading_score + students_df.writing_score
```

```python
students_df = students_df.assign(total_score_log = np.log(students_df.total_score))
```

```python
concentrations = pd.read_csv('http://stepik.org/media/attachments/course/4852/algae.csv')
```

```python
mean_concentrations = (
    concentrations.groupby('genus', as_index=False)
    .agg({'sucrose': 'mean',
          'alanin': 'mean',
          'citrate': 'mean',
          'glucose': 'mean',
          'oleic_acid': 'mean'
         })
    .rename(columns={
        'sucrose': 'sucrose_mean',
        'alanin': 'alanin_mean',
        'citrate': 'citrate_mean',
        'glucose': 'glucose_mean',
        'oleic_acid': 'oleic_acid_mean'
    })
    
)
```

```python
concentrations.head()
```

```python
round((
    concentrations[concentrations.genus == 'Fucus']
    .alanin
    .describe()), 2
)
```

```python
(
    concentrations
    .groupby('genus')
    .agg({'alanin':
          ['min', 'mean', 'max']})
    .loc['Fucus'].round(2)
)
```

```python
(
    concentrations
    .groupby('group')
    .agg({'species': 'count',
         'citrate': 'var',
         'sucrose': ['max','min']
         })
)
```
